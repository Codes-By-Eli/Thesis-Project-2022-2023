```python
import os
import email

'''
This cell grabs searches through the directory of .eml files and converts them into string.
Some emails had to be deleted because my computer ascertained them to be too dangerous to open or was of wrong typing.

Create try catch that deletes the emails that are unable to open
This cell returns two arrays.
    1. The content of the email
    2. The file name
'''


#This is for creating the array file names and contents for Phishing Emails
def listDir(dir):

    file_names = []
    texts = []
    fileNames = os.listdir(dir)

    for fle in fileNames:        
        try:
            msg = email.message_from_file(open(dir + '\\' +fle))
            message = msg.as_string()
            texts.append(message)
            name = dir+ '\\' +fle
            file = name.split('_phishing\\')
            file_names.append(file[1])
        except:
            print(dir + '\\'+ fle)
           
    return texts, file_names

#This is for creating the array file names and contents for Ham Emails
def listDir2(dir):

    file_names = []
    texts = []
    fileNames = os.listdir(dir)

    for fle in fileNames:        
        try:
            msg = email.message_from_file(open(dir + '\\' +fle))
            message = msg.as_string()
            texts.append(message)
            name = dir+ '\\' +fle
            file = name.split('ham\\')
            file_names.append(file[1])
        except:
            print(dir + '\\'+ fle)
           
    return texts, file_names



folderPath = r'C:\Users\imyug\OneDrive - Iona College\Honors Classes\HON401\Petrovic Work\Datasets\emails(phishing)\public_phishing'
folderPath2 = r'C:\Users\imyug\OneDrive - Iona College\Honors Classes\HON401\Petrovic Work\Datasets\ham'

files = ['20051114',  
        'phishing0', 
        'phishing2', 
        'phishing3']

contents = []
fileNames = []        

for i in files:
    newFolderPath = folderPath + '\\' + i
    currContents, currFileNames = listDir(newFolderPath)
    for j in range(len(currContents)):
        contents.append(currContents[j])
        fileNames.append(currFileNames[j])


hamContents, hamFileNames = listDir2(folderPath2)
for l in range(len(hamFileNames)):
    fileNames.append(hamFileNames[l])
print(len(fileNames))
```

```python
'''
Work with one file to try and extract keywords
Use NLP library to assist

Features:
1. # of </a>
2. # of https
3. # of http
4. Most frequent words (Around 100 or so)

'''
#----These are the necessary libraries for counting and cleansing-----------
import nltk
from bs4 import BeautifulSoup
import lxml

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.stem.porter import PorterStemmer

import matplotlib.pyplot as plt
%matplotlib inline
#-----------------------------------------------------------------------

def cleanhtml(html):
    soup = BeautifulSoup(html)
    text = soup.get_text()
    return text

def countElements(words):
    words_no_punc = []

    for w in words:
        if w.isalpha():
            words_no_punc.append(w.lower())



    clean_words = []
    for w in words_no_punc:
        if w not in stopwords:
            clean_words.append(w)


    stem_words = []
    for w in clean_words:
        stem_words.append(porter.stem(w))

    
    countAccount = 0
    countInformation = 0
    countUpdate = 0
    countPlease = 0
    countSecure = 0
    countAlert = 0
    countRequest = 0
    countClick = 0
    countFradulent = 0
    countNotification = 0
    countUpgrade = 0
    countIndef = 0
    countAccess = 0
    countPassword = 0
    countVerify = 0
    countProvide = 0
    countConfidential = 0

    
    for el in stem_words:
        if el == 'account':
            countAccount += 1
        elif el == 'inform':
            countInformation += 1
        elif el == 'updat':
            countUpdate += 1
        elif el == 'pleas':
            countPlease += 1
        elif el == 'secur':
            countSecure += 1
        elif el == 'alert':
            countAlert += 1
        elif el == 'request':
            countRequest += 1
        elif el == 'click':
            countClick += 1
        elif el == 'fraudul':
            countFradulent += 1
        elif el == 'notif':
            countNotification += 1
        elif el == 'upgrad':
            countUpgrade += 1
        elif el == 'indefinit':
            countIndef += 1
        elif el == 'access':
            countAccess += 1
        elif el == 'password':
            countPassword += 1
        elif el == 'verif':
            countVerify += 1
        elif el == 'provid':
            countProvide += 1
        elif el == 'confidenti':
            countConfidential += 1


    accountCount.append(countAccount)
    informationCount.append(countInformation)
    updateCount.append(countUpdate)
    pleaseCount.append(countPlease)
    secureCount.append(countSecure)
    alertCount.append(countAlert)
    requestCount.append(countRequest)
    clickCount.append(countClick)
    fradulentCount.append(countFradulent)
    notificationCount.append(countNotification)
    upgradeCount.append(countUpgrade)
   
    indefiniteCount.append(countIndef)
    accessCount.append(countAccess)
    passwordCount.append(countPassword)
    verifyCount.append(countVerify)
    provideCount.append(countProvide)
    confidentialCount.append(countConfidential)


stopwords = stopwords.words("english")
porter = PorterStemmer()

commonWords =[]

#-----This will store the count of the words in each email-----------------
accountCount = []
informationCount = []
updateCount = []
pleaseCount = []
secureCount = []
alertCount = []
requestCount = []
clickCount = []
fradulentCount = []
notificationCount = []
upgradeCount = []
indefiniteCount = []
accessCount = []
passwordCount = []
verifyCount = []
provideCount = []
confidentialCount = []
hamOrPhish = []
#--------------------------------------------------------------------


#--------This will perform the counting of the key words------------
for i in range(len(contents)):
    convert = contents[i].lower()
    message = convert.split('x-keywords:')
    noTags = cleanhtml(message[1])
    words = word_tokenize(noTags)

    countElements(words)
    
    hamOrPhish.append(1)

    #fdist = FreqDist(stem_words)

    #commonWords.append(fdist.most_common(30))

for i in range(len(hamContents)):
    convert = hamContents[i].lower()
    message = convert.split('date:')
    noTags = cleanhtml(message[1])
    words = word_tokenize(noTags)

    countElements(words)
    
    hamOrPhish.append(0)

    #fdist = FreqDist(stem_words)

    #commonWords.append(fdist.most_common(30))
```
